## 7. Zusammenführung aller „Versionen“ einer Variablen zu einer einzigen

Auswahl konsistenter Variablen:
Beim Zusammenführen der Jahresdateien zeigte sich, dass viele Variablen je nach Jahr unterschiedliche Namen oder Strukturen aufweisen.
Um einen konsistenten, vergleichbaren und modellierbaren Datensatz zu erhalten, wurden nur jene Variablen beibehalten, die in allen Jahren vorhanden sind und inhaltlich dieselben Konzepte abbilden.
Dadurch werden strukturelle NaN‑Werte vermieden und eine saubere Grundlage für Regression, Clustering und die spätere Dashboard‑Erstellung geschaffen.

```python
### 7.1. Erstellung eine Funktion zum Zusammenführen von Spalten
```

```python
def merge_columns(df, col_list):
    series = None
    for col in col_list:
        if col in df.columns:
            if series is None:
                series = df[col].copy()
            else:
                series = series.fillna(df[col])
    return series
```

```python
### 7.2 Vereinigung von Spalten
```

```python
# Überprüfung der Namen

df.columns
```

```python
# Land

df["country_final"] = merge_columns(df, [
    "country",
    "country_or_region",
    "country_name"
])
```

```python
# Region

df["region_final"] = merge_columns(df, [
    "regional_indicator",
    "region"
])
```

```python
[col for col in df.columns if "happiness" in col]
```

```python
# Happiness Score

df["happiness_final"] = merge_columns(df, [
    "ladder_score",
    "happiness_score",
    "score",
    "happiness.score"
])
```

```python
# GDR

df["gdp_final"] = merge_columns(df, [
    "gdp_per_capita",
    "economy_(gdp_per_capita)",
    "economy..gdp.per.capita.",
    "logged_gdp_per_capita",
    "explained_by:_log_gdp_per_capita",
    "explained_by:_gdp_per_capita"
])
```

```python
# Gesundheit

df["health_final"] = merge_columns(df, [
    "health_(life_expectancy)",
    "health..life.expectancy.",
    "health..life.expectancy"
    "explained_by:_healthy_life_expectancy"

])
```

```python
# Social Support

df["social_support_final"] = merge_columns(df, [
    "family",
    "social_support",
    "explained_by:_social_support"
])
```

```python
# Freiheit

df["freedom_final"] = merge_columns(df, [
    "freedom_to_make_life_choices",
    "freedom",
    "explained_by:_freedom_to_make_life_choices"
])
```

```python
# Corruption

df["corruption_final"] = merge_columns(df, [
    "perceptions_of_corruption",
    "trust_(government_corruption)",
    "trust..government.corruption.",
    "explained_by:_perceptions_of_corruption"
])
```

### 7.3. Finaly Datensatz

```python
columns_to_keep = [
    "country_final",
    "region_final",
    "happiness_final",
    "gdp_final",
    "social_support",
    "health_final",
    "freedom_final",
    "generosity",
    "corruption_final"
]

df_clean = df[columns_to_keep].copy()
```
